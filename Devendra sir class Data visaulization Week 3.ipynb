```python
import numpy as np
from scipy import stats
```

# Making scatter plot

```python
import matplotlib.pyplot as plt
x=10*np.random.randn(10)
yerr=10*np.random.randn(10)
y=3*x+3+yerr
plt.scatter(x,y)
```

# fitting the line with scipy

```python
def myfunc(x):
    return slope*x+intercept
```

```python
fig,ax=plt.subplots(figsize=(10,5))
slope,intercept,r,p,std_err=stats.linregress(x,y)
mymodel=list(map(myfunc,x))
ax.errorbar(x,y,yerr=std_err,fmt='o',capsize=5)
ax.plot(x,mymodel,c='r',linewidth=2)
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.tick_params(direction='in');
```

Decay Law

```python
A0=50
t=np.linspace(0,74,10)
A=A0*np.exp(-8.693*t/74)
yerr=10*np.random.random_sample(10)
plt.errorbar(t,A,yerr=yerr,fmt='-o',capsize=5)
plt.fill_between(t,A-yerr,A+yerr,color='gray', alpha=0.2)
plt.xlabel('Time (Year)')
plt.ylabel('Activity (Curie)')
plt.title('Decay Scheme');
```
